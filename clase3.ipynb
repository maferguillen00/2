```python
my_pivot = pd.pivot_table(df,
                          index = ['STNAME', 'CTYNAME'],
                          values = ['BIRTHS2010', 'BIRTHS2011'],
                          #aggfunc = ['std', 'sum', 'mean', 'median']
                          aggfunc = 'median',
                          columns = 'SUMLEV'
                         )
my_pivot
```

```python
df = pd.read_csv('datasets/airline_data.csv',low_memory=False)
df.head()
```

```python
df.shape
```

```python
columz = df.columns.tolist()
columz.sort()
columz
```

```python
delay_pivot = pd.pivot_table(df,
                             index = ['Year', 'Quarter'],
                             values = 'DepDelay',
                             aggfunc = 'sum',
                             columns = 'DepartureDelayGroups',
                             margins = True
                            )
delay_pivot
```

```python
df[['DepDelay','DepartureDelayGroups']]
```
